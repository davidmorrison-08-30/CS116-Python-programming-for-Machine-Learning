# **IMPORT NECESSARY LIBRARIES**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import drive 
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from wordcloud import WordCloud, STOPWORDS
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score
import nltk, re
```

```python
drive.mount("/content/gdrive")
```

# **DATASET**

```python
dataset = pd.read_csv("gdrive/My Drive/Restaurant_Reviews.tsv", delimiter = '\t', quoting = 3)
```

```python
nltk.download('stopwords')

n_sample = len(dataset)
corpus = []
for i in range(n_sample):
  scru = re.sub("[^a-zA-Z]", " ", dataset["Review"][i])
  scru = scru.lower().split()
  whole_stopwords = stopwords.words("english")
  whole_stopwords.remove("not")
  scru = [PorterStemmer().stem(word) for word in scru if not word in set(whole_stopwords)]
  scru = " ".join(scru)
  corpus.append(scru)
```

```python
print(corpus)
```

```python
def create_wordcloud(list_sample):
  comment_words = ''
  stopwords = set(STOPWORDS)   

  tokens = []
  for i in list_sample:
      tokens.append(i.lower())
      comment_words += " ".join(tokens)+" "
  
  wordcloud = WordCloud(width = 800, height = 800,
                  background_color ='white',
                  stopwords = stopwords,
                  min_font_size = 10).generate(comment_words)
  plt.figure(figsize = (4, 4), facecolor = None)
  plt.imshow(wordcloud, interpolation='bilinear')
  plt.axis("off")
  plt.tight_layout(pad = 0)

  plt.show()
```

```python
create_wordcloud(corpus)
```

# **INITIATE A BAG-OF-WORDS MODEL**

```python
cv = CountVectorizer(max_features=1000)
X = cv.fit_transform(corpus).toarray()
y = dataset.iloc[:, 1].values
```

# **SPLIT DATASET**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=0)
```

# **TRAIN A NAIVE BAYES MODEL**

```python
GNB = GaussianNB()
GNB.fit(X_train, y_train)
```

```python
y_pred = GNB.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=GNB.classes_)
disp.plot()
```
